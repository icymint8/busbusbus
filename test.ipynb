```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss  
    
data= [
    166,
    415,
    590,
    394,
    1099,
    479,
    288,
    49,
    669,
    305,
    232,
    461,
    934,
    174,
    514,
    134,
    669,
    113,
    101,
    176,
    1146,
    518,
    319,
    100,
    211,
    97,
    419,
    219,
    1
]       

chi_linespace = np.linspace(-100, 1300, len(data))
x, y, z = ss.chi2.fit(data)
pdf_chi2 = ss.chi2.pdf(chi_linespace, x,y,z)

plt.hist(data, bins=15, density=True, alpha=0.3)
plt.hist(np.random.default_rng().chisquare(x, len(data)) * z + y, bins=10, density=True, alpha=0.3)
plt.plot(chi_linespace, pdf_chi2, label='Chi2')
plt.legend()
plt.show()
```

```python
chi_linespace = np.linspace(18120, 20000, 1000)
plt.plot(chi_linespace, ss.expon.pdf(chi_linespace, *(18120.0, 259.375)), label='exp')
plt.hist(np.random.default_rng().exponential(259.375, 1000) + 18120.0, bins=100, density=True, alpha=0.3)
plt.legend()
plt.show()
```

```python
chi_linespace = np.linspace(17500, 20000, 1000)
plt.plot(chi_linespace, ss.t.pdf(chi_linespace, *(213485839.75521147, 18566.862108442438, 268.92220358051526)), label='t')
plt.hist(np.random.default_rng().standard_t(213485839.75521147, 1000) * 268.92220358051526 + 18566.862108442438, bins=100, density=True, alpha=0.3)
plt.legend()
plt.show()
```

```python
chi_linespace = np.linspace(18000, 18350, 1000)
plt.plot(chi_linespace, ss.norm.pdf(chi_linespace, *(18146.714285714286, 34.97229223956869)), label='norm')
plt.hist(np.random.default_rng().normal(*(18146.714285714286, 34.97229223956869), 1000), bins=100, density=True, alpha=0.3)
plt.legend()
plt.show()
```

```python
chi_linespace = np.linspace(-100, 2200, 1000)
plt.plot(chi_linespace, ss.gamma.pdf(chi_linespace, *(16.98237077685359, -673.7165670743836, 81.58805945154745)), label='gamma')
plt.hist(np.random.default_rng().gamma(16.98237077685359, 81.58805945154745, 1000) - 673.7165670743836, bins=100, density=True, alpha=0.3)
plt.legend()
plt.show()
```

```python
chi_linespace = np.linspace(500, 1000, 100)
plt.plot(chi_linespace, ss.beta.pdf(chi_linespace, *(1.5848929430742282, 0.7955861513566723, 541.7394666657118, 377.26053333428825)), label='beta')
plt.hist(np.random.default_rng().beta(1.5848929430742282, 0.7955861513566723, 1000) * 377.26053333428825 + 541.7394666657118, bins=100, density=True, alpha=0.3)
plt.legend()
plt.show()
```

```python
# scale, location 둘 다 신경써야 되는 놈: beta, chi2, t
# location만 신경쓰면 되는 놈: gamma, exponential
# 다 때려넣으면 되는 놈: normal
```
